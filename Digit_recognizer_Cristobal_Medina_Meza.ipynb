Define X, y and X_test

```python
imageid = test.index

y = train.iloc[:, 0]
X = train.iloc[::, 1:]
X = X.drop([0])
y = y.drop([0])

X_test = test
X_test = X_test.drop([0])
```

```python
X
```

Scaler

```python
#The pixels change from 0 to 255
X = X.astype(int) / 255
X_test = X_test.astype(int) / 255
```

Split X for training and validating

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X,y, test_size = 0.1, random_state=101)
X_train.shape, X_val.shape, y_train.shape, y_val.shape, X_test.shape
```

```python
from traitlets.traitlets import default
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(50, ), max_iter=50, alpha=1e-4, solver='sgd', verbose=10,random_state=54,learning_rate='adaptive',).fit(X_train,y_train)
print(model.score(X_train,y_train))
```

```python
from traitlets.traitlets import default
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(50, ), max_iter=1000, alpha=1e-4, verbose=10,random_state=54,learning_rate='adaptive',).fit(X_train,y_train)
print(model.score(X_train,y_train))
```

```python
y_pred = model.predict(X_val)
print(model.score(X_val,y_val))
```

```python
df_val = pd.DataFrame({'Real_value': y_val, 'Pred._value': y_pred})
df_val.reset_index(drop=True, inplace=True)
df_val
```
